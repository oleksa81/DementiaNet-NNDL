```python
import numpy as np
import pandas as pd
from pathlib import Path
from tqdm import tqdm
from sklearn.model_selection import train_test_split

from PIL import Image  # for image validation

from google.colab import drive
drive.mount('/content/drive')

# Path setup
save_path = Path('/content/drive/MyDrive/DementiaNetProject/data')

dm_path = save_path / 'dementia'
nd_path = save_path / 'nodementia'

# Supported image formats
IMG_EXTS = ['*.png', '*.jpg', '*.jpeg']

data = []

# --- Load dementia images ---
for ext in IMG_EXTS:
    for path in tqdm(dm_path.glob(ext)):
        try:
            img = Image.open(path)  # verify image loads
            data.append({
                'file': path.stem,
                'label': 'dementia',
                'path': path
            })
        except Exception as e:
            print(f"{path} invalid image:", e)

# --- Load non-dementia images ---
for ext in IMG_EXTS:
    for path in tqdm(nd_path.glob(ext)):
        try:
            img = Image.open(path)
            data.append({
                'file': path.stem,
                'label': 'nodementia',
                'path': path
            })
        except Exception as e:
            print(f"{path} invalid image:", e)

# Convert to DataFrame
df = pd.DataFrame(data)

print("Total samples:", len(df))
print(df['label'].value_counts())

# --- Train/Validation Split ---
train_df, valid_df = train_test_split(
    df,
    test_size=0.2,
    stratify=df['label'],  # keeps class balance
    random_state=42
)

# Reset index
train_df = train_df.reset_index(drop=True)
valid_df = valid_df.reset_index(drop=True)

# Save CSVs (optional, for later training)
train_df.to_csv(save_path / 'train_img.csv', sep='\t', index=False)
valid_df.to_csv(save_path / 'valid_img.csv', sep='\t', index=False)

print(f"Train: {len(train_df)}")
print(f"Valid: {len(valid_df)}")
```
